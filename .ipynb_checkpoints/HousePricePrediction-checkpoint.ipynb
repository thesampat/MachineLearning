## LinearRegression without scaling data

```python
# lets fill null value
nul_col = [col for col in df.columns if df[col].isnull().sum()>=1]
```

```python
df = df.fillna(method='bfill')
```

```python
df = pd.get_dummies(df, drop_first = True)
```

```python
df.head()
```

```python
x = df.drop('SalePrice', axis =1)
y = df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state =42)
```

```python
from sklearn.linear_model import LinearRegression
linearm = LinearRegression()
linearm.fit(x_train, y_train)
linearm.score(x_test, y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
dirr = DecisionTreeRegressor()
dirr.fit(x_train, y_train)
dirr.score(x_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators = 200)
rfr.fit(x_train, y_train)
rfr.score(x_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knnr = KNeighborsRegressor()
knnr.fit(x_train, y_train)
knnr.score(x_test, y_test)
```

```python
from sklearn.ensemble import AdaBoostRegressor
abr = AdaBoostRegressor(learning_rate = 0.44, n_estimators = 100)
abr.fit(x_train, y_train)
abr.score(x_test, y_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr = GradientBoostingRegressor(learning_rate = 0.1)
gbr.fit(x_train, y_train)
gbr.score(x_test, y_test)
```
