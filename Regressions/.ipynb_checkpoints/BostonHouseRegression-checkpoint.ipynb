```python
# import necessary libraries
import pandas as pd
import numpy as np
import sklearn
```

```python
# imprt boston sklearn data
from sklearn.datasets import load_boston
data = load_boston()
```

```python
# create pandas DataFarme
sub_df = pd.DataFrame(data=data.data, columns=data.feature_names)
```

```python
# concating both feature and target data
df = pd.concat([sub_df, pd.Series(data.target, name='PRICE')], axis=1)
```

```python
# chech null vaues
df.isnull().sum()
```



AS WE CAN SEE THERE IS NO NULL VAUES

```python
# lets split data feature and target
x = df.drop('PRICE', axis =1)
y = df['PRICE']
```

```python
# now make train and test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 42)
```

# Simple Linear Regression

```python
# model training
from sklearn.linear_model import LinearRegression
linearmodel = LinearRegression()
linearmodel.fit(x_train, y_train)
linearmodel.score(x_train, y_train)
```

We got 0.73 as a train data accuracy 

```python
pred = linearmodel.predict(x_test)
```

```python
# lets check predicted accuracy
from sklearn.metrics import r2_score, mean_squared_error, accuracy_score
a = r2_score(y_test, pred)
b = mean_squared_error(y_test, pred)
a, b
```

# Lasso Regression
